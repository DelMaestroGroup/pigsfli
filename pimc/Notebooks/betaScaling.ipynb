# Relative Errors vs $\beta$

```python
# Load exact values
S2_SF_exact = 1.2760231964505537
S2_PT_exact = 0.861757259131563
S2_MOTT_exact = 0.2018374508123394
```

```python
# Load PIMC estimates for various betas
S2_SF_PIMC = np.array(S2_SF_PIMC)
S2_PT_PIMC = np.array(S2_PT_PIMC)
S2_MOTT_PIMC = np.array(S2_MOTT_PIMC)

S2_SF_PIMC_err = np.array(S2_SF_PIMC_err)
S2_PT_PIMC_err = np.array(S2_PT_PIMC_err)
S2_MOTT_PIMC_err = np.array(S2_MOTT_PIMC_err)
```

```python
# Compute relative errors
S2_SF_relErr = (np.abs(S2_SF_PIMC - S2_SF_exact)/S2_SF_exact) * 100
S2_PT_relErr = (np.abs(S2_PT_PIMC - S2_PT_exact)/S2_PT_exact) * 100
S2_MOTT_relErr = (np.abs(S2_MOTT_PIMC - S2_MOTT_exact)/S2_MOTT_exact) * 100
```

```python
# Compute the error propagation
S2_SF_PIMC_err_err = S2_SF_PIMC_err / S2_SF_exact 
S2_PT_PIMC_err_err = S2_PT_PIMC_err / S2_PT_exact
S2_MOTT_PIMC_err_err = S2_MOTT_PIMC_err / S2_MOTT_exact
```

```python
fig,ax1 = plt.subplots()
ax1.errorbar(betas[1:],S2_MOTT_relErr[1:], yerr=S2_MOTT_PIMC_err_err[1:], fmt='.', capsize=3,ms=0)
ax1.scatter(betas[1:],S2_MOTT_relErr[1:],marker='s',lw=0.47,label=r'$U=10.0$',color=colors[0],ec='black',zorder=10)
ax1.plot(betas[1:],S2_MOTT_relErr[1:],lw=1,color=colors[0])

ax1.errorbar(betas[1:],S2_PT_relErr[1:], yerr=S2_PT_PIMC_err_err[1:], fmt='.', capsize=3,ms=0)
ax1.scatter(betas[1:],S2_PT_relErr[1:],marker='^',lw=0.47,label=r'$U=3.3$',color=colors[1],ec='black',zorder=10)
ax1.plot(betas[1:],S2_PT_relErr[1:],lw=1,color=colors[1])

ax1.errorbar(betas[1:],S2_SF_relErr[1:], yerr=S2_SF_PIMC_err_err[1:], fmt='.', capsize=3,ms=0)
ax1.scatter(betas[1:],S2_SF_relErr[1:],marker='o',lw=0.47,label=r'$U=0.1$',color=colors[2],ec='black',zorder=10)
ax1.plot(betas[1:],S2_SF_relErr[1:],lw=1,color=colors[2])

ax1.set_ylabel(r'$\% \; \rm{Error} \; S_2$')
ax1.set_xlabel(r'$\beta$')
ax1.annotate(r"$D=1,L=4,N=4,\ell_{\rm{max}}=2$",(0.42,0.32),xycoords='figure fraction')
# ax1.annotate(r"$\beta=4.0$",(0.71,0.45),xycoords='figure fraction')
# ax1.annotate(r"$\ell_{\rm{max}}=2$",(0.71,0.45),xycoords='figure fraction')
ax1.axhline(0,color="#888888",zorder=-1)
ax1.set_xlim(betas[1:][0])
# ax1.set_ylim(-0.05)
ax1.tick_params(direction='in')

plt.legend()
plt.savefig("relativeErrors.pdf",dpi=400)
plt.savefig("relativeErrors.png",dpi=400)
plt.plot()
```
