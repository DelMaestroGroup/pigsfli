# Relative Errors vs $\beta$

```python
# Load exact values
S2_SF_exact = 1.2760231964505537
S2_PT_exact = 0.861757259131563
S2_MOTT_exact = 0.2018374508123394
```

```python
# Load PIMC estimates for various betas
S2_SF_PIMC = np.array([1.51580689,1.47423826,1.41899132,1.36231594,1.31254398,1.28124215,1.27906441,1.27543449,1.27657474,1.27648126])
S2_PT_PIMC = np.array([1.36923716,1.23583285,1.10742763,0.98918430,0.90621856,0.87466592,0.86479341,0.86190217,0.86185875,0.86082436,0.86421632])
S2_MOTT_PIMC = np.array([1.14703756,0.90666066,0.59729885,0.33174670,0.22518995,0.20538175,0.20242293,0.20216403,0.20157087,0.20179480,0.20249579])

S2_SF_PIMC_err = np.array([0.00030638,0.00030130,0.00084902,0.00082654,0.00077185,0.00105125,0.00116583,0.00159049,0.00173492,0.00177282])
S2_PT_PIMC_err = np.array([0.00040320,0.00034489,0.00077856,0.00060761,0.00055925,0.00082570,0.00064116,0.00097075,0.00080298,0.00104466,0.00115528])
S2_MOTT_PIMC_err = np.array([0.00031146,0.00027290,0.00051550,0.00041323,0.00035551,0.00041519,0.00037153,0.00041070,0.00056387,0.00056599,0.00054289])
```

```python
# Compute relative errors
S2_SF_relErr = (np.abs(S2_SF_PIMC - S2_SF_exact)/S2_SF_exact) * 100
S2_PT_relErr = (np.abs(S2_PT_PIMC - S2_PT_exact)/S2_PT_exact) * 100
S2_MOTT_relErr = (np.abs(S2_MOTT_PIMC - S2_MOTT_exact)/S2_MOTT_exact) * 100
```

```python
# Compute the error propagation
S2_SF_PIMC_err_err = S2_SF_PIMC_err / S2_SF_exact 
S2_PT_PIMC_err_err = S2_PT_PIMC_err / S2_PT_exact
S2_MOTT_PIMC_err_err = S2_MOTT_PIMC_err / S2_MOTT_exact
```
